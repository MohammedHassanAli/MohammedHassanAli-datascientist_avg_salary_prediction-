```python
#Feature Engineering 
#function that it could simplify the job title
def title_simplifier(title_complete):
    found = False
    job_titles = ['data scientist' , 'data engineer' , 'analyst' ,'machine learning' , 'manager' , 'director']
    for title in job_titles:
        if title in  title_complete.lower():
            found = True
            return title
    
    if found == False:
        return 'na'
        
   
```

```python
def seniority(title):
    lwr_title = title.lower()
    if 'sr' in lwr_title or 'senior' in lwr_title or 'lead' in lwr_title or 'principal' in lwr_title:
        return 'senior'
    elif 'jr' in lwr_title:
        return 'junior'
    else:
        return 'na'
```

```python
df['job_simp'] = df['Job Title'].apply(title_simplifier)
```

```python
df.job_simp.value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(seniority)
```
